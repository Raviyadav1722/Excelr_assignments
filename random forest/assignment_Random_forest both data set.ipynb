```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets  
import numpy as np
from sklearn import preprocessing 
```

# Random forest classification for company_data set

```python
cmp=pd.read_csv('Company_Data.csv')
cmp
```

```python
cmp.dtypes
```

```python
cmp["sales"]="<=7.5"
cmp.loc[cmp["Sales"]>=7.5,"sales"]="High Sales"
cmp.loc[cmp["Sales"]<=7.5,"sales"]="Low Sales"
cmp
```

```python
cmp1=cmp.iloc[:,1:]
cmp1
```

```python
label_encoder = preprocessing.LabelEncoder()
cmp1['ShelveLoc']= label_encoder.fit_transform(cmp['ShelveLoc']) 
cmp1['Urban']= label_encoder.fit_transform(cmp['Urban']) 
cmp1['US']= label_encoder.fit_transform(cmp['US']) 
cmp1['sales']= label_encoder.fit_transform(cmp['sales']) 
cmp1
```

```python
# Random Forest Classification
from pandas import read_csv
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier

x=cmp1.iloc[:,0:10]
y=cmp1.iloc[:,10]
num_trees = 100
max_features = 3
kfold = KFold(n_splits=10,random_state=None)
model = RandomForestClassifier(n_estimators=num_trees, max_features=max_features)
results = cross_val_score(model, x, y, cv=kfold)
print(results.mean()) 
```

# Random forest classification for fraud_check data set

```python
frd=pd.read_csv('Fraud_check.csv')
frd
```

```python
frd.info()
```

```python
frd["tax_inc"]= "<=30000"
frd.loc[frd["Taxable.Income"]>=30000,"tax_inc"]="good"
frd.loc[frd["Taxable.Income"]<=30000,"tax_inc"]="risky"
frd
```

```python
frd1=frd.drop(columns="Taxable.Income")
frd1
```

```python
label_encoder = preprocessing.LabelEncoder()
frd1['Undergrad']= label_encoder.fit_transform(frd1['Undergrad']) 
frd1['Urban']= label_encoder.fit_transform(frd1['Urban']) 
frd1['Marital.Status']= label_encoder.fit_transform(frd1['Marital.Status']) 
frd1['tax_inc']= label_encoder.fit_transform(frd1['tax_inc']) 
frd1
```

```python
# Random Forest Classification
from pandas import read_csv
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier

x=frd1.iloc[:,0:5]
y=frd1.iloc[:,5]
num_trees = 100
max_features = 3
kfold = KFold(n_splits=10,random_state=None)
model = RandomForestClassifier(n_estimators=num_trees, max_features=max_features)
results = cross_val_score(model, x, y, cv=kfold)
print(results.mean()) 
```
